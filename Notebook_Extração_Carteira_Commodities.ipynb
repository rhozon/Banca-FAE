```python
# Importa as libs do Python

import pandas as pd
from yahooquery import Ticker
```

```python

# Lista de tickers das commodities (minha carteira)
TICKERS = [
    "ZC=F",  # Corn Futures
    "ZO=F",  # Wheat Futures
    "KE=F",  # KC HRW Wheat Futures
    "ZR=F",  # Rough Rice Futures
    "GF=F",  # Feeder Cattle Futures
    "ZS=F",  # SoyMeal Futures
    "ZM=F",  # Soybean Meal Futures
    "ZL=F"   # SoyBeans Futures
]

```

```python

# Cria objeto Ticker para as commodities
commodities = Ticker(TICKERS)

# Baixa o histórico de preços em um intervalo específico
# Ajuste conforme a necessidade, por exemplo "start" e "end"
# ou use "period='1y'", "period='max'", etc.
df = commodities.history(period='1y')

# O DataFrame tem índices (symbol, date). Vamos reindexar para transformar em colunas
df = df.reset_index()

# Filtra somente as colunas de interesse: data, símbolo e fechamento
df_close = df[['date', 'symbol', 'close']]

# Transformamos para formato 'tabela pivô', com cada commodity em uma coluna
# e cada linha representando a data.
df_pivot = df_close.pivot(index='date', columns='symbol', values='close')

# Visualizar as primeiras linhas
display(df_pivot.head())
```


## Salva em .csv os dados das cotações da minha carteira


```python

# Salva em CSV (apenas valores de fechamento)
df_pivot.to_csv('commodities_close_prices.csv', index=True)
print("CSV gerado com sucesso: commodities_close_prices.csv")
```
